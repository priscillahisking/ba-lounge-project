```python
import pandas as pd
df= pd.read_excel(r'C:\Users\Surface Laptop Go\Downloads\British Airways Summer Schedule Dataset - Forage Data Science Task 1.xlsx')
def assign_tiers(row):
    if row['HAUL']=='LONG':
        return pd.Series({'Tier1%' : 5 , 'Tier2%': 15 , 'Tier3%' : 30})
    else:
        return pd.Series({ 'Tier1%' :1 , 'Tier2%' : 5 , 'Tier3%' :20})
tier_estimates = df.apply(assign_tiers,axis=1)

lookup_table =pd.concat([
    df[['FLIGHT_NO' , 'FLIGHT_TIME','ARRIVAL_STATION_CD','HAUL']].rename(columns={
        'FLIGHT_NO':'Flight No',
        'FLIGHT_TIME': 'Departure Time',
        'ARRIVAL_STATION_CD':'Destination',
        'HAUL': 'Category'
    }),
    tier_estimates
],axis =1)
lookup_table.head()
lookup_table.to_csv('ba_lounge_lookup_table.csv', index=False)
```
